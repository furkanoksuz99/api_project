```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
import re

def scrape_hepsiemlak(base_url, max_batch=3):
    # Chrome ayarları
    options = Options()
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--window-size=1920,1080")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)

    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

    all_data = []

    # Şehir bilgisini linkten al
    city_match = re.search(r"https://www\.hepsiemlak\.com/([\w-]+)-satilik", base_url)
    city = city_match.group(1).capitalize() if city_match else "Bilinmiyor"

    # İlk 5 sayfa
    for page in range(1, 6):
        url = f"{base_url}?page={page}"
        driver.get(url)
        time.sleep(3)

        soup = BeautifulSoup(driver.page_source, "html.parser")
        listings = soup.find_all("li", class_="listing-item")

        print(f"Sayfa {page}: {len(listings)} ilan bulundu.")
        for item in listings:
            title_tag = item.find("h3")
            title = title_tag.text.strip() if title_tag else "Başlık yok"
            link_tag = item.find("a", class_="card-link")
            link = "https://www.hepsiemlak.com" + link_tag["href"] if link_tag else "Link yok"
            all_data.append({
                "city": city,
                "title": title,
                "link": link
            })

    # Sonra 40'ar 40 artarak ilerle
    start_page = 41
    batch_count = 0
    while batch_count < max_batch:
        end_page = start_page + 39
        for page in range(start_page, end_page + 1):
            url = f"{base_url}?page={page}"
            driver.get(url)
            time.sleep(3)

            soup = BeautifulSoup(driver.page_source, "html.parser")
            listings = soup.find_all("li", class_="listing-item")

            if not listings:
                print(f"Sayfa {page}: İlan yok, durduruluyor.")
                driver.quit()
                return all_data

            print(f"Sayfa {page}: {len(listings)} ilan bulundu.")
            for item in listings:
                title_tag = item.find("h3")
                title = title_tag.text.strip() if title_tag else "Başlık yok"
                link_tag = item.find("a", class_="card-link")
                link = "https://www.hepsiemlak.com" + link_tag["href"] if link_tag else "Link yok"
                all_data.append({
                    "city": city,
                    "title": title,
                    "link": link
                })

        start_page += 40
        batch_count += 1

    driver.quit()
    return all_data

# Kullanım
url = "https://www.hepsiemlak.com/ankara-satilik"
results = scrape_hepsiemlak(url, max_batch=1)  # max_batch=1 demek 41-80 arası 1 batch tarar

# Sonuçları yazdır
for idx, r in enumerate(results, 1):
    print(f"{idx}. [{r['city']}] {r['title']} \n   {r['link']}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time

def parse_hepsiemlak_page(url):
    options = Options()
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--window-size=1920,1080")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

    driver.get(url)
    time.sleep(3)  # sayfanın tam yüklenmesini bekle

    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()

    listings = soup.find_all("li", class_="listing-item")
    data = []

    for item in listings:
        title_tag = item.find("h3")
        title = title_tag.text.strip() if title_tag else ""

        link_tag = item.find("a", class_="card-link")
        link = "https://www.hepsiemlak.com" + link_tag["href"] if link_tag else ""

        price_tag = item.find("span", class_="list-view-price")
        price = price_tag.text.strip() if price_tag else ""

        date_tag = item.find("span", class_="list-view-date")
        date = date_tag.text.strip() if date_tag else ""

        housing_type_tag = item.find("span", class_="short-property")
        housing_type = housing_type_tag.text.strip() if housing_type_tag else ""

        celly_spans = item.find_all("span", class_="celly")
        room_count = ""
        sqm = ""
        building_age = ""
        floor = ""

        for span in celly_spans:
            text = span.text.strip()
            if "+" in text and " " in text:
                room_count = text
            elif "m²" in text:
                sqm = text
            elif "Yaşında" in text:
                building_age = text
            elif "Kat" in text:
                floor = text

        location_tag = item.find("span", class_="list-view-location")
        location_text = location_tag.text.strip() if location_tag else ""
        city, district, neighborhood = ("", "", "")
        if location_text:
            parts = [part.strip() for part in location_text.split("/")]
            if len(parts) >= 3:
                city, district, neighborhood = parts[0], parts[1], parts[2]
            elif len(parts) == 2:
                city, district = parts[0], parts[1]
            elif len(parts) == 1:
                city = parts[0]

        data.append({
            "şehir": city,
            "ilçe": district,
            "mahalle": neighborhood,
            "başlık": title,
            "link": link,
            "konut_tipi": housing_type,
            "oda_sayısı": room_count,
            "metrekare": sqm,
            "bina_yaşı": building_age,
            "kat": floor,
            "fiyat": price,
            "ilan_tarihi": date
        })

    return data

def scrape_multiple_pages(city_name, max_pages=100, pause_every=30, pause_seconds=3):
    all_data = []
    base_url = f"https://www.hepsiemlak.com/{city_name}-satilik?page="

    for page in range(1, max_pages + 1):
        print(f"Sayfa {page} çekiliyor...")
        url = base_url + str(page)
        page_data = parse_hepsiemlak_page(url)
        print(f"Sayfa {page}: {len(page_data)} ilan bulundu.")
        all_data.extend(page_data)

        if page % pause_every == 0:
            print(f"{pause_every} sayfa tamamlandı. {pause_seconds} saniye bekleniyor...")
            time.sleep(pause_seconds)

    return all_data

# Örnek kullanım
city = "ankara"  # Linkteki şehir ismi küçük harf
max_pages = 30   # 30 sayfa çek
ilanlar = scrape_multiple_pages(city, max_pages=max_pages)

# Örnek çıktı
for i, ilan in enumerate(ilanlar, 1):
    print(f"{i}. {ilan}\n")
```
